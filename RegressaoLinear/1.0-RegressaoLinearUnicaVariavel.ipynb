```python
# Importação das Bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scr.utils import title_labels,gradienteDescendenteUnicaVariavel

plt.rcParams["figure.figsize"] = (10,6)
```

```python
# Importando os Dados Utilizados
caminhoData1 = r'data/raw/arquivos/data1.txt'
dadosFoodTruck = pd.read_csv(caminhoData1,names=['Populacao','Lucro'])
```

```python
plt.scatter(x=dadosFoodTruck['Populacao'],y=dadosFoodTruck['Lucro'],c='green')
title_labels('Lucro x População',14,'População','Lucro')
```

```python
valoresX = dadosFoodTruck['Populacao']
valoresY = dadosFoodTruck['Lucro']
valorA = 1
valorB = 0.5
```

```python
valora,valorb,erroQuadradoMedio = gradienteDescendenteUnicaVariavel(valoresX,valoresY,valorA,valorB,800)
```

```python
erroQuadradoMedio
```

```python
plt.plot(list(range(0,len(erroQuadradoMedio))),erroQuadradoMedio)
title_labels('Erro Quadrado Medio x Iterações',14,'Iterações','Erro Quadrado Medio')
plt.show()
```

```python
regression_line = [(valora*x) + valorb for x in dadosFoodTruck['Populacao']]

plt.figure(figsize=(10, 7))
plt.scatter(dadosFoodTruck.Populacao, dadosFoodTruck.Lucro, color='g')
plt.plot(dadosFoodTruck.Populacao, regression_line, color='b')
plt.text(x=12,y=22,s=f'f(x) = {round(valora,2)}*x {round(valorb,4)}',color='b')
title_labels('Lucro x População',14,'População','Lucro')
plt.legend(["Curva Regressao Linear","Dados"])
plt.grid()
plt.show()
```
