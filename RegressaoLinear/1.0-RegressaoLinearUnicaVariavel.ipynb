```python
# Importação das Bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scr.utils import title_labels,gradienteDescendente
```

```python
# Importando os Dados Utilizados
caminhoData1 = r'data/raw/arquivos/data1.txt'
dadosFoodTruck = pd.read_csv(caminhoData1,names=['Populacao','Lucro'])
```

```python
plt.scatter(x=dadosFoodTruck['Populacao'],y=dadosFoodTruck['Lucro'],c='green')
title_labels('Lucro x População',14,'População','Lucro')
```

```python
valoresX = dadosFoodTruck['Populacao']
valoresY = dadosFoodTruck['Lucro']
valorA = 7
valorB = 1
```

```python
valora,valorb,erroQuadradoMedio = gradienteDescendente(valoresX,valoresY,valorA,valorB,800)
```

```python
plt.plot(erroQuadradoMedio)
title_labels('Erro Quadrado Medio x Iterações',14,'Iterações','Erro Quadrado Medio')
```

```python
regression_line = [(valora*x) + valorb for x in dadosFoodTruck['Populacao']]

plt.figure(figsize=(10, 7))
plt.scatter(dadosFoodTruck.Populacao, dadosFoodTruck.Lucro, color='g')
plt.plot(dadosFoodTruck.Populacao, regression_line, color='b')
plt.text(x=12,y=22,s=f'f(x) = {round(valora,2)}*x {round(valorb,4)}',color='b')
title_labels('Lucro x População',14,'População','Lucro')
plt.grid()
plt.show()
```

```python
teste = dadosFoodTruck
teste.columns = ['Grade','Salary']
```

```python
df = teste
```

```python
from matplotlib import pyplot as plt
import pandas as pd

m = 7
b = 1
learning_rate = 0.000001

for i in range(800):
  y_pred = m*valoresX + b
  m_derivative = sum(2*valoresX*(y_pred - valoresY))
  b_derivative = sum(2*(y_pred - valoresY))
  m = m - (learning_rate * m_derivative)
  b = b - (learning_rate * b_derivative)
  # print(m, b) # Remove comments to view step-by-step.

print("\nAngular Coefficient (m): {0}\nLinear Coefficient (b): {1}".format(round(m), round(b)))

regression_line = [(m*x) + b for x in dadosFoodTruck['Populacao']]

plt.figure(figsize=(10, 7))
plt.scatter(dadosFoodTruck.Populacao, df.Salary, color='g')
plt.plot(dadosFoodTruck.Populacao, regression_line, color='b')
plt.title('Grades vs Salaries | Gradient descent Approach')
plt.xlabel('Grade')
plt.ylabel('Salary')
plt.grid()
#plt.savefig('../images/plot-03.png', format='png')
plt.show()
```
